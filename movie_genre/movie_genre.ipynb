```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ---Data loading and inspection---
# Load the dataset
file_name= 'movie_genre_classification_final.csv'
df = pd.read_csv(file_name)
```

```python
#  Display the first 5 rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Display information about the DataFrame
print("\nInformation about the DataFrame:")
print(df.info())
```

```python
# ---Data Cleaning and Preprocessing  ----
# Check for duplicates and drop them if any
print("\nNumber of duplicate rows:",df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("\nNumber of rows after dropping duplicates:", len(df))
```

```python
# ---Exploratory Data Analysis( EDA) ---
# Calculate descriptive statistics for numerical columns
numerical_cols =['Year', 'Duration', 'Rating', 'Votes', 'Budget_USD', 'BoxOffice_USD', 'Num_Awards', 'Critic_Reviews']
print("\nDescriptive statistics for numerical columns:")
print(df[numerical_cols].describe().T.to_markdown(numalign ="left", stralign="left"))
```

```python
# Find value counts for 'Genre' (top 10)
print("\nValue counts for 'Genre' (top 10):")
print(df['Genre'].value_counts().head(10).to_markdown(numalign="left" , stralign="left"))

# Find value counts for 'Director' (top 10)
print("\nValue counts for 'Director' (top 10):")
print(df['Director'].value_counts().head(10).to_markdown(numalign="left" , stralign="left"))

# Find value counts for 'Language' (top 10)
print("\nValue counts for 'Language' (top 10):")
print(df['Language'].value_counts().head(10).to_markdown(numalign="left", stralign="left"))

# Find value counts for 'Country' (top 10)
print("\nValue counts for 'Country' (top 10):")
print(df['Country'].value_counts().head(10).to_markdown(numalign="left", stralign="left"))

# Find value counts for 'Content_Rating' (top 10)
print("\nValue counts for 'Content_Rating' (top 10):")
print(df['Content_Rating'].value_counts().head(10).to_markdown(numalign="left", stralign="left"))

# Find value counts for 'Lead_Actor' (top 10)
print("\nValue counts for 'Lead_Actor' (top 10):")
print(df['Lead_Actor'].value_counts().head(10).to_markdown(numalign="left", stralign="left"))
```

```python
# ---Visualization ---
# Create a histogram for the 'Rating' distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True, color='purple')
plt.title('Distribution of Movie Ratings', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('rating_distribution.png')
```

```python
# Create a bar chart for the top 10 genres
top_genres = df['Genre'].value_counts().head(10).sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.index, y=top_genres.values, palette='plasma')
plt.title('Top 10 Most Frequent Movie Genres', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_genres.png')
```

```python
# Create a boxplot for Rating by Content_Rating
plt.figure(figsize=(10, 7))
sns.boxplot(x='Content_Rating', y='Rating', data=df, order=df['Content_Rating'].value_counts().index, palette='coolwarm')
plt.title('Distribution of Ratings by Content Rating', fontsize=16)
plt.xlabel('Content Rating', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('rating_by_content_rating.png')
```

```python
# Create a scatter plot of Budget_USD vs. BoxOffice_USD
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Budget_USD', y='BoxOffice_USD', data=df, alpha=0.6, color='darkgreen')
plt.title('Relationship between Budget and Box Office Revenue', fontsize=16)
plt.xlabel('Budget (USD)', fontsize=12)
plt.ylabel('Box Office (USD)', fontsize=12)
plt.grid(linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('budget_vs_boxoffice.png')
```
